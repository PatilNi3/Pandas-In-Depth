13. Join

df.join() method is used to combine two DataFrames based on their indices. This method is particularly useful for merging DataFrames that share a common index or when you want to add columns from one DataFrame to another.

df_1.join(df_2, how='left', on=None)

    df_2    - The DataFrame to join with.
    how     - The type of join (default is 'left')
            - 'left': Use the index from the df_1 DataFrame.
            - 'right': Use the index from the df_2 DataFrame.
            - 'outer': Use the union of the indices from both DataFrame.
            - 'inner': Use the intersection of the indices from both DataFrame.
    on      - Column or index level names to join on; must be found in both DataFrames.



```python
df9 = pd.DataFrame(
    {
        'A':['A1', 'A2', 'A3', 'A4', 'A5'],
        'B':['B1', 'B2', 'B3', 'B4', 'B5'],
        'C':['C1', 'C2', 'C3', 'C4', 'C5'],
        'index':[1, 2, 3, 4, 5]
    }
    )

df10 = pd.DataFrame(
    {
        'D':['D1', 'D2', 'D3', 'D4', 'D5'],
        'E':['E1', 'E2', 'E3', 'E4', 'E5'],
        'F':['F1', 'F2', 'F3', 'F4', 'F5'],
        'index':[4, 5, 6, 7, 8]
    }
    )
```

```python
df9.join(df10, how='left')
```

20. Scaling or Feature Scaling

Feature scaling is an essential preprocessing step in machine learning that involves adjusting the range of feature values. It helps algorithms converge faster and improves model performance, especially for distance-based algorithms like k-nearest neighbors and gradient descent.

1. Normalization (Min-Max Scaling)

Normalization scales data to a specific range, often 0 & 1. This is done by subtracting min value from the feature and dividing by range (max-min).

Normalization = [df - df.min()] / [df.max() - df.min()]

```python
df8
```

```python
Normalization = (df8['age'] - df8['age'].min())/(df8['age'].max() - df8['age'].min())
Normalization
```

2. Standardization (Z-Score Scsling)

Standardization transforms data to have a mean of 0 and standard deviation of 1. This is achieved by subtracting the mean of each feature from the data and then dividing by the standard deviation.
In dataset during performing EDA operations we need to convert different different units of different different features into one unit, so for that we use standardization.

Standardization = [df - df.mean()] / [df.std()]

```python
Standardization = (df8['age'] - df8['age'].mean())/(df8['age'].std())
Standardization
```

3. Robust Scaling

Robust Scaling can be achieved by subtracting median of each feature from the data and then dividing by the Inter Quartile Range.
It is robust to outliers.

Robust Scaling = [df - df.median()] / IQR

```python
Robust_Scaling = (df8['age'] - df8['age'].median())/(df8['age'].quantile(0.75)-df8['age'].quantile(0.25))
Robust_Scaling
```
